```python
import os
import torch
import tensorflow as tf
from transformers import BertModel, BertTokenizer
```

```python
# load the pretrained weights v1.1 1m pubmed
# the dropout are disabled when load from pretrained weights
biobert = BertModel.from_pretrained("./checkpoints/biobert_v1.1_pubmed_torch/")
```

```python
# load the vocab
bbtokenizer_args = {
    "vocab_file": "./checkpoints/biobert_v1.1_pubmed_torch/vocab.txt",
    "do_lower_case": False,
}
bbtokenizer = BertTokenizer(**bbtokenizer_args)
```

```python
s = torch.tensor([bbtokenizer.encode("hello!")])
print(s)
```

```python
biobert(s)
```

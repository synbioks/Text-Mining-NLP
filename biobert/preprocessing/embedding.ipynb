```python
import os
import torch
import tensorflow as tf
from transformers import BertModel, BertTokenizer
```

```python
# get all the txt files
txt_path = os.path.abspath("../acs-data/txt-files")
txt_files = os.listdir(txt_path)
txt_files = list(map(lambda x: os.path.join(txt_path, x), txt_files))
print("%d text files" % len(txt_files))
```

```python
# split txt file into sentences
def to_sentences(filename):
    lines = []
    with open(filename, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            print(len(line.split()))
            lines.append(line)
    print(lines)
```

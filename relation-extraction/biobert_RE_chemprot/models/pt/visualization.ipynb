```python
def heat_map(data):
    data_max = np.amax(data)
    data_min = np.amin(data)
    data_range = data_max - data_min
    print(f"min: {data_min}, max: {data_max}")
    data = (data - data_min) / data_range
    fig, ax = plt.subplots(figsize=(10,4))
    im = ax.imshow(data)
    fig.colorbar(im)
    plt.show()
```

```python
ACT_PATH = "activations/top-model-9000"
with open(os.path.join(ACT_PATH, "l0.npy"), "rb") as fin:
    l0 = np.load(fin)
with open(os.path.join(ACT_PATH, "l1.npy"), "rb") as fin:
    l1 = np.load(fin)
with open(os.path.join(ACT_PATH, "l2.npy"), "rb") as fin:
    l2 = np.load(fin)
heat_map(np.sort(np.var(l0, axis=0)).reshape(16, 32))
heat_map(np.sort(np.var(l1, axis=0)).reshape(16, 32))
heat_map(np.var(l2, axis=0).reshape(1, 4))
```
